```python
#Importamos las librerías a utilizar
from google.colab import drive
import pandas as pd
import numpy as np

#Solicitamos permisos y montamos la carpeta de Google Drive  
drive.mount('/content/drive')

#Leemos el archivo
movies = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DataSets/datos_ejercicios/medias.csv', low_memory=False)

#Determinamos el 20% de las películas con más votos para establecer el mínimo de votos aceptados
votes_filter= movies['conteo'].quantile(0.80)

#Seleccionamos solo películas que tengan el  mínimo de votos aceptado
movies_filter = movies[movies.conteo >= votes_filter].reset_index()

#Calculamos el promedio del rating de las películas seleccionadas
media = movies_filter.media_rating.mean()

#Definimos una función con la fórmula a usar para el ranking
def media_RME(df, m=votes_filter, C=media):
    v = df['conteo']
    R = df['media_rating']
    # Calculamos el weighted score
    return round((R*v + C*m) / (v+m),2)

#Aplicamos la función a todo el dataset y creamos una nueva columna para almacenar el resultado
movies_filter["Puntuación"] = movies_filter.apply(media_RME,axis = 1)

#Establecemos un orden descendente  
movies_filter = movies_filter.sort_values('Puntuación', ascending=False)

#Mostramos las mejores 25 películas y renombramos los nombres de las columnas.
movies_filter[['imdb_title', 'conteo', 'media_rating', 'Puntuación']].rename(
   columns={'imdb_title':'Título', 'conteo':'Cantidad de Votos','media_rating':'Promedio de Votos'}).head(25)
```
